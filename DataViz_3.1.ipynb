```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x, y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
```

```python
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

```python
plt.plot(x, y, color='#ff0000', marker='o', markersize=10, linestyle='--')

plt.xlim(0, 2)
plt.ylim(100, 300)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```
